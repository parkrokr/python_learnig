```python
import requests # 요청하다
from bs4 import BeautifulSoup # 파이썬으로 HTML 을 다루는 라이브러리
import pandas as pd # 엑셀화
import time

url = requests.get("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EB%A1%9C%EB%98%90")

html = BeautifulSoup(url.text)

numbers = html.find('div',class_ ='num_box').find_all('span')
numbers

del numbers[6] # 보너스 번호 삭제

box = []

for i in numbers:
  box.append(int(i.text)) # 문자열로 되어있어서 정수로 변환

box
```

```python
current = int(html.find('a', class_ = '_lotto-btn-current').find('em').text.replace('회', ''))
current
```

```python
total = []

for n in range(1, current + 1):
  url = requests.get(f"https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query={n}%ED%9A%8C+%EB%A1%9C%EB%98%90")
  html = BeautifulSoup(url.text)

  numbers = html.find('div',class_ ='num_box').find_all('span')

  del numbers[6] # 보너스 번호 삭제

  box = []

  for i in numbers:
   box.append(int(i.text)) # 문자열로 되어있어서 정수로 변환

  total.append(box)

  print(f'{n}회차 로또 데이터 저장완료 : {box}')

  time.sleep(0/3)
```

```python
total

df = pd.DataFrame(total,columns = ['1번째', '2번째', '3번째', '4번째', '5번째', '6번째', '보너스번호'])
df.to_excel('lotto.xlsx')
df
```

```python
#실습 : 연금복권 데이터를 1회차부터 현재 회차까지 크롤링 해서 엑샐화까지

url = requests.get("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EC%97%B0%EA%B8%89%EB%B3%B5%EA%B6%8C")

html = BeautifulSoup(url.text)

current = int(html.find('a', class_ = '_lottery-btn-current').find('em').text.replace('회', ''))

total = []

for n in range(1, current + 1):
  url = requests.get(f"https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query={n}%ED%9A%8C%EC%97%B0%EA%B8%88%EB%B3%B5%EA%B6%8C")
  html = BeautifulSoup(url.text)

  numbers = html.find('ul',class_ = 'win_num').find_all('span')

  box = []

  for i in numbers:
    box.append(int(i.text[0]))

  
  total.append(box)

  print(f'{n}회차 연급복권 데이터 저장완료 : {box}')

  time.sleep(0/3)

```

```python
total

df = pd.DataFrame(total,columns = ['조', '1번째', '2번째', '3번째', '4번째', '5번째', '6번째'])
df.to_excel('result.xlsx')
df
```

```python
url = requests.get("https://finance.naver.com/sise/sise_market_sum.nhn?sosok=0&page=1")
html = BeautifulSoup(url.text)

last_page = int(html.find('td', class_ = 'pgRR').find('a')['href'][-2:])

kospi_box = []

for n in range(1, last_page + 1):

  url = requests.get(f"https://finance.naver.com/sise/sise_market_sum.nhn?sosok=0&page={n}")
  html = BeautifulSoup(url.text)
  table = html.find('table', class_ = 'type_2')

  table = pd.read_html(str(table))[0]
  table = table[table['종목명'].notnull()]

  del table['N']
  del table['토론실']

  kospi_box.append(table)
```

```python
len(kospi_box)
```

```python
kospi = pd.concat(kospi_box, ignore_index=True)
kospi['종류'] = ['KOSPI'] * len(kospi)
kospi.to_excel('kospi.xlsx')
kospi

```

```python
# 실습 : 코스닥도 모두 크롤링해서 코스피랑 합쳐서 엑셀화

from tqdm import tqdm

url = requests.get("https://finance.naver.com/sise/sise_market_sum.nhn?sosok=1&page=1")
html = BeautifulSoup(url.text)

last_page = int(html.find('td', class_ = 'pgRR').find('a')['href'][-2:])

kosdaq_box = []

for n in tqdm(range(1, last_page + 1)):

  url = requests.get(f"https://finance.naver.com/sise/sise_market_sum.nhn?sosok=1&page={n}")
  html = BeautifulSoup(url.text)
  table = html.find('table', class_ = 'type_2')

  table = pd.read_html(str(table))[0]
  table = table[table['종목명'].notnull()]

  del table['N']
  del table['토론실']

  kosdaq_box.append(table)
  
kosdaq = pd.concat(kosdak_box, ignore_index=True)
kosdaq['종류'] = ['KOSDAQ'] * len(kosdaq)
kosdaq.to_excel('kosdaq.xlsx')
kosdaq
```
