```python
import seaborn as sns

anscombe = sns.load_dataset("anscombe")
anscombe

anscombe[anscombe['dataset'] == 'I']
```

```python
data1 = anscombe[anscombe['dataset'] == 'I']
data2 = anscombe[anscombe['dataset'] == 'II']
data3 = anscombe[anscombe['dataset'] == 'III']
data4 = anscombe[anscombe['dataset'] == 'IV']
```

```python
# 평균
print(data1.mean())
print(data2.mean())
print(data3.mean())
print(data4.mean())
```

```python
#표준편차
print(data1.std())
print(data2.std())
print(data3.std())
print(data4.std())
```

```python
# 누락값 처리하기

# 빈칸을 0 으로 처리하기
# 빈칸을 평균값으로 처리하기
# 빈칸이 있는 행을 없애버리기
# 주변에 있는 값으로 채우기

```

```python
ebola.fillna(0) # NaN 값을 원하는 값(0) 으로 채움
```

```python
ebola['Cases_Guinea'].mean()
ebola['Cases_Guinea'].fillna(ebola['Cases_Guinea'].mean())
# 평균 넣으니 갑자기 값이 확 튐
# 앞 뒤 값 평균이 적절해 보임
```

```python
ebola.interpolate() # 앞에있는거 뒤에있는거의 평균으로 누락값 처리( 대신 앞이나 뒷값이 없으면 평균을 못내기 떄문에 처리를 못함)
```

```python
ebola = ebola.interpolate()
ebola
```

```python
ebola.fillna(method = 'bfill') # back fill
```

```python
ebola = ebola.fillna(method = 'bfill')
```

```python
ebola.count(0)
```

```python
ebola = ebola.fillna(method='ffill') # front fill
ebola
```
