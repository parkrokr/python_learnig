```python
# 지도 학습

# y^ = ax + b
# 오차 = 시그마 (y^ - y)^2 = 시그마(ax + b - y)^2

# 미분 : a' = 2(ax + b -y)x
#        b' = 2(ax + b -y)x

# a = a - a'
# b = b - b'



# 기울기 8~ 10, 절편 20 ~ 40
import numpy as np
import matplotlib.pyplot as plt


a = 10
b = 30

x = np.array([2,4,6,8])
y = np.array([40,85,75,95])

plt.scatter(x,y)
plt.plot(x, a* x + b,'r')
```

```python
a = 10
b = 30

x = np.array([2,4,6,8])
y = np.array([40,85,75,95])

lr = 0.005  # 학습률

for i in range(3000):
  
  y_hat = a * x + b
  error = y_hat - y
  
  #미분
  a_diff = sum(2 * error * x)
  b_diff = sum(2 * error * 1)

  #덮어쓰기
  a = a - a_diff * lr
  b = b - b_diff * lr

  print(f'{i+1}회 학습 // 기울기 : {a} // 절편 : {b}')
```

```python
plt.scatter(x,y)
plt.plot(x, a* x + b,'r')
```

```python
a1 = 0
a2 = 0
b = 0
lr = 0.005

x1 = np.array([2,4,6,8]) # 공부시간
x2 = np.array([1,2,1,3]) # 문제집 푼 개수
y = np.array([40,85,75,95])

lr = 0.005  # 학습률

for i in range(4000):
  
  y_hat = a1 * x1 + a2 * x2 + b
  error = y_hat - y
  
  #미분
  a1_diff = sum(2 * error * x1)
  a2_diff = sum(2 * error * x2)
  b_diff = sum(2 * error * 1)

  #덮어쓰기
  a1 = a1-a1_diff *lr
  a2 = a2 - a2_diff * lr
  b = b - b_diff * lr

  print(f'{i+1}회 학습 // 기울기1 : {a1} // 기울기2 : {a2}//절편 : {b}')
```

```python
plt.scatter(x,y)
plt.plot(x, a1 * x1 + a2 * x2 + b, 'r')\

# a1,a2 : 가중치
```

```python
# 평균 사과 사진 // 평균 파인애플 사진 // 평균 바나나 사진

# 300개의 사진에서 사과의 평균 사진 빼기 -> 사과의 평균 데이터가 빠진 300개의 사진

#데이터가 0 에 가까울수록 사과 사진

abs_diff = abs(fruits - apple_mean)
abs_mean = np.mean(abs_diff, axis = (1,2))

apple_index = np.argsort(abs_mean)[:100]

fig, axs = plt.subplots(10, 10, figsize= (10,10))

for i in range(10):
  for j in range(10):
    axs[i, j].imshow(fruits[apple_index[i*10 + j]], cmap = 'gray_r')
    axs[i, j].axis('off')

plt.show()
```

```python
#실습
import pandas as pd
wine = pd.read_csv("https://bit.ly/wine-date")
wine
```

```python
# 0 : 레드와인 // 1이면 화이트화인

data = wine[['alcohol','sugar', 'pH']].to_numpy()
target = wine[['class']].to_numpy()

```

```python
# 훈련 데이터/ 시험데이터 나누기
# 데이터 표준화
# 모델 생성
# 학습
# 평가
```

```python
# 훈련 데이터/ 시험데이터 나누기
train_input, test_input, train_target, test_target = train_test_split(data, target)
```

```python
# 데이터 표준화
mean = np.mean(train_input, axis= 0)
std = np.std(train_input, axis = 0)

train_scaled = (train_input - mean)/ std
test_scaled = (test_input - mean)/ std
```

```python
# 모델 생성
kn = KNeighborsClassifier()

#학습
kn.fit(train_scaled, train_target)
```

```python
#평가
kn.score(test_scaled, test_target)
```
