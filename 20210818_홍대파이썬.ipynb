```python
# 데이터 크롤링 : 인터넷 데이터를 가져오는 기술

# HTML

import requests # 요청하다
from bs4 import BeautifulSoup # 파이썬으로 HTML 을 다루는 라이브러리
import pandas as pd # 엑셀화
import time

# 정확하겐 크롤링이 아니라 스크랩핑

# 크롤링 : 다 가져오는것
# 스크랩핑 : 부분적을 가져오는것

#가져말지 말라고 적혀있는걸 가져오면 불법

#데이터를 너무 빨리 가져오면 들킴(트래픽 증가)


#200 : 정상
#400대 : 없는 페이지
#500대 : 문제가 있거나 거절당함
url = requests.get("https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EB%A1%9C%EB%98%90")
#url.text # HTML 처럼 생긴 문자열

html = BeautifulSoup(url.text) # 이 문자열을 HTML 로 번역


# find ; HTML 로 반환해줌
# find_all : 리스트로 반환해줌


# tag : dic, 클래스 속성값 : num_box 
# span 이란것을 다 찾겠다
numbers = html.find('div',class_ ='num_box').find_all('span')
numbers

del numbers[6] # 보너스 번호 삭제

box = []

for i in numbers:
  box.append(int(i.text)) # 문자열로 되어있어서 정수로 변환

box
```

```python
# 현제 회차
current = int(html.find('a', class_ = '_lotto-btn-current').find('em').text.replace('회', ''))
current
```

```python
for n in range(1, current + 1):
  url = requests.get(f"https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query={n}%ED%9A%8C+%EB%A1%9C%EB%98%90")
  html = BeautifulSoup(url.text)

  numbers = html.find('div',class_ ='num_box').find_all('span')

  del numbers[6] # 보너스 번호 삭제

  box = []

  for i in numbers:
   box.append(int(i.text)) # 문자열로 되어있어서 정수로 변환

  print(box)

  time.sleep(1)
```
