```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
import time

titles = []
summary = []

for n in tqdm(range(1,3992, 10)):

  url = requests.get(f"https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%ED%99%8D%EC%9D%B5%EB%8C%80&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=41&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start={n}")
  html = BeautifulSoup(url.text)


  for i in html.find('ul', class_ = 'list_news').find_all('li',class_ = 'bx'):
    titles.append(i.find('a',class_='news_tit').text)
    summary.append(i.find('div',class_ = 'news_dsc').text)

  time.sleep(0.01)
```

```python
t = ''

for i in titles:
  t += i
  t += ' '

for i in summary:
  t += i
  t += ' '

len(t)
```

```python
# 정규표현식 : 원하는 패턴의 문자열을 가장 효율적으로 찾는 방법

import re

tel = 'hello 1234567890 world'

# 연속된 10자리 숫자

# \d : 숫자
# \D : 숫자가 아닌거

# \s : 공백
# \S : 공백이 아닌거

# \w : 문자
# \W : 문자가 아닌거

# ? : 있을수도 있고 없을수도 있다
# () : 그 부분만 catch 한다

# \w{3,7} : 문자가 3개 이상 7까지 연속됨
# \w+ : 문자가 1개 이상
# \w* : 문자가 0개 이상

# . : 모든것
# [] : 포함
# [a-zA-Z] : 영어
# [가-힣] : 한글

#  ^ : 시작
# $ : 끝

# https://regexone.com/ 연습 사이트

re.findall('\d\d\d\d\d\d\d\d\d\d', tel) # tel 이란 변수에서 숫자 10개의 패턴을 찾고 싶다 - > 패턴을 찾아서 리스트에 넣어줌
```

```python
tel = 'hello 123 456 7890 world'
re.findall('\d\d\d\s\d\d\d\s\d\d\d\d',tel)
```

```python
re.findall('\d{3}\s\d{3}\s\d{4}',tel) # 이렇게도 가능함
```

```python
# 전화번호 밑에 두개다 잡아야 하는 경우
# 01012345678
# 010 1234 5678


# ? : 있을수도 있고 없을수도 있다
re.findall('\d{3}\s?\d{3}\s?\d{4}',tel)
```

```python
# () : 그 부분만 catch 한다
tel = '(123) 456 7890'

re.findall('\(\d{3}\)\s?\d{3}\s?\d{4}',tel)
```

```python
tel = '+1 (123) 456 7890'

re.findall('\+\d\s\(\d{3}\)\s?\d{3}\s?\d{4}',tel)
```

```python
# 크롤링

box = []
for i in re.findall('[가-힣]+',t):
  if len(i) > 1:
    box.append(i)

# 빈도분석
dic = {}
for i in box:
  if i not in dic:
    dic[i] = 1
  else:
    dic[i] += 1
  
dic
```

```python
import pandas as pd

df = pd.DataFrame({
    '단어' : list(dic.keys()),
    '빈도수' : list(dic.values())
})

df.sort_values(by = '빈도수', ascending= False)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt


wc = WordCloud(font_path = 'BMDOHYEON_ttf.ttf', background_color = 'white').generate_from_frequencies(dic)

plt.figure(figsize = (10,8))
plt.axis('off')
plt.imshow(wc)
plt.show()
```
